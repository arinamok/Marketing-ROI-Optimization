```python
import pandas as pd
```

```python
# Load full dataset
df = pd.read_csv('marketing_AB_enhanced.csv')
# Print the first 5 rows
print(df.head())
```

```python
# Drop unnamed column
df = df.loc[:, ~df.columns.str.contains('^unnamed', case=False)]
```

```python
# Print the first 5 rows
print(df.head())
```

```python
# Randomly sample 35,000 rows
df_sample = df.sample(n=35000, random_state=42)
```

```python
# Rename columns to snake_case
df_sample.columns = ['user_id', 'test_group', 'converted', 'total_ads',
                     'most_ads_day', 'most_ads_hour', 'revenue',
                     'cost_per_impression', 'total_cost', 'profit', 'roi']
```

```python
# Check first 3 rows
print(df_sample.head(3))
```

```python
# Save sample for SQL import
df_sample.to_csv('marketing_AB_sample.csv', index=False)
```
